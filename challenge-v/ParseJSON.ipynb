```python
import json
import pandas as pd

filename = ('./cloud-conformity-homepage.json')

with open(filename) as f:
    data = json.load(f)
    
print(data.keys())
    
flat_article_list = []
i = 0;
for group in data['data']:
    i+=1
    article_list = group['relationships']['rules']['data']
    for article in article_list:
        articles = {
            'group_name' : group['id'],
            'article_id' : article['id']
        }
        flat_article_list.append(articles)

flat_attribute_list = []
for article_desc in data['included']:
    print(article_desc['id'])
    attributes = article_desc['attributes']
    attribute_format = {}
    for attribute in attributes:
        attribute_format[attribute] = attributes[attribute]
    attribute_format['article_id'] = article_desc['id']
    flat_attribute_list.append(attribute_format)

att = pd.DataFrame.from_records(flat_attribute_list)
art = pd.DataFrame.from_records(flat_article_list)
print i
```

```python
result = pd.merge(art,
                 att,
                 on='article_id', 
                 how='left')

result
```
